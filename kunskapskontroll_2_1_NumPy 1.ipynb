##### Remark on Broadcasting when doing Linear Algebra calculations in Python. 
From the mathematical rules of matrix addition, the operation below (m1 + m2) does not make sense. The reason is that matrix addition requires two matrices of the same size. In Python however, it works due to broadcasting rules in NumPy. So you must be careful when doing Linear Algebra calculations in Python since they do not follow the "mathematical rules". This can however easily be handled by doing some simple programming, for example validating that two matrices have the same shape is easy if you for instance want to add two matrices.

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([1, 1])
print(m1 + m2)
```

### Vector and matrix algebra

Now you are going to create a function that can be reused every time you add or multiply matrices. The function is created so that we do the addition and multiplication according to the rules of vector- and matrix algebra.

Create a function "add_mult_matrices" that takes two matrices as input arguments (validate that the input are of the type numpy.ndarray by using the isinstance function), a third argument that is either 'add' or 'multiply' that specifies if you want to add or multiply the matrices (validate that the third argument is either 'add' or 'multiply'). When doing matrix addition, validate that the matrices have the same size. When doing matrix multiplication, validate that the sizes conform (i.e. number of columns in the first matrix is equal to the number of rows in the second matrix).

```python
def add_mult_matrices(matrix1, matrix2, operation):
    """
    Addera eller multiplicera två matriser enligt linjär algebras regler.
    
    Parameters:
    matrix1, matrix2: numpy arrays
    operation: 'add' eller 'multiply'
    
    Returns:
    Resultatet av operationen
    """
    # Validera att inputs är numpy arrays
    if not isinstance(matrix1, np.ndarray):
        raise TypeError("matrix1 måste vara en numpy.ndarray")
    if not isinstance(matrix2, np.ndarray):
        raise TypeError("matrix2 måste vara en numpy.ndarray")
    
    # Validera operation
    if operation not in ['add', 'multiply']:
        raise ValueError("operation måste vara antingen 'add' eller 'multiply'")
    
    if operation == 'add':
        # För addition måste matriserna ha samma shape
        if matrix1.shape != matrix2.shape:
            raise ValueError(f"Matriserna måste ha samma shape för addition. Fick {matrix1.shape} och {matrix2.shape}")
        return matrix1 + matrix2
    
    elif operation == 'multiply':
        # För multiplikation måste kolumner i första matrisen = rader i andra matrisen
        if matrix1.shape[1] != matrix2.shape[0]:
            raise ValueError(f"För multiplikation måste kolumner i första matrisen ({matrix1.shape[1]}) vara lika med rader i andra matrisen ({matrix2.shape[0]})")
        return matrix1 @ matrix2

# Testa funktionen
print("Test 1 - Addition:")
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(add_mult_matrices(A, B, 'add'))

print("\nTest 2 - Multiplikation:")
C = np.array([[1, 2], [3, 4]])
D = np.array([[2, 0], [1, 3]])
print(add_mult_matrices(C, D, 'multiply'))

print("\nTest 3 - Felhantering (olika shapes för addition):")
try:
    E = np.array([[1, 2]])
    F = np.array([[1], [2]])
    add_mult_matrices(E, F, 'add')
except ValueError as e:
    print(f"Fel: {e}")
```

```python
# Hjälpfunktion för vektornorm
def vector_norm(v):
    """Beräkna normen (längden) av en vektor"""
    return np.sqrt(sum([xi**2 for xi in v]))

# Uppgift 10.1.1
x = (4, 3)

print("=== Uppgift 10.1.1 ===")
print("dim x:", len(x))
print("5x:", [5*xi for xi in x])
print("3x:", [3*xi for xi in x])
print("5x + 3x:", [5*xi + 3*xi for xi in x])
print("8x:", [8*xi for xi in x])
print("4x - x:", [4*xi - xi for xi in x])
print("||x||:", vector_norm(x))
print()

# Uppgift 10.1.2
v = (3, 7, 0, 11)

print("=== Uppgift 10.1.2 ===")
print("dim v:", len(v))
print("2v:", [2*vi for vi in v])
print("5v + 2v:", [5*vi + 2*vi for vi in v])
print("4v - 2v:", [4*vi - 2*vi for vi in v])
print("||v||:", vector_norm(v))
print()

# Uppgift 10.1.3
v1 = (4, 3, 1, 5)
v2 = (2, 3, 1, 1)

print("=== Uppgift 10.1.3 ===")
print("||v1||:", vector_norm(v1))

diff = [a - b for a, b in zip(v1, v2)]
print("v1 - v2:", diff)
print("||v1 - v2||:", vector_norm(diff))
```

### Copies and Views
Read the link https://numpy.org/doc/stable/user/basics.copies.html before starting the exercises below. 

Basic indexing creates a view. How can you check if v1 and v2 is a view or copy? If you change the last element in v2 to 123, will the last element in v1 be changed? Why?

```python
v1 = np.arange(4)
v2 = v1[-2:]
print(v1)
print(v2)
```

```python
# Kontrollera om v2 är en view eller kopia med 'base'-attributet
print("Är v2 en view? Kontrollera om v2.base är v1:", v2.base is v1)
print()

# Ändra sista elementet i v2
print("Ändrar sista elementet i v2 till 123...")
v2[-1] = 123

print("\nEfter modifiering:")
print("v1:", v1)
print("v2:", v2)
print()

# Svar: Ja, v1 kommer att ändras!
# Varför? Eftersom v2 är en VIEW (inte en kopia) av v1.
# När man slicear med basic indexing (v1[-2:]), skapar NumPy en view
# som delar samma minne som v1. Alla ändringar i v2 påverkar v1.
# Detta är effektivt för stora arrayer, men var försiktig med oavsiktliga ändringar!

print("Förklaring:")
print("v2 är en VIEW av v1, så de delar samma minne.")
print("Ändringar i v2 återspeglas i v1.")
```
